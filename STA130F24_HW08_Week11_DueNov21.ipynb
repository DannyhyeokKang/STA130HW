### 2. Continue your ChatBot session and explore with your ChatBot what real-world application scenario(s) might be most appropriately addressed by each of the following *metrics* below: provide your answers and, in your own words, *concisely explain your rationale for your answers.*<br>


1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1. Accuracy

•	Spam Detection: When false positives (marking legitimate emails as spam) and false negatives (missing spam emails) are equally problematic.

•	Quality Control in Manufacturing: Evaluating whether a product is defect-free or faulty.

2. Sensitivity

•	Medical Diagnosis: Detecting diseases like cancer, where failing to identify a positive case (FN) can have severe consequences.

•	Search and Rescue Operations: Identifying missing individuals, where missing a true case (FN) could result in loss of life.

3. Specificity

•	Fraud Detection: Ensuring legitimate transactions are not flagged as fraudulent to avoid inconvenience or financial loss.

•	Drug Approval Testing: Avoiding the approval of ineffective or harmful drugs by correctly identifying non-effective trials (true negatives).

4. Percision

•	Spam Email Detection: Avoiding misclassification of legitimate emails as spam, which could cause important emails to be missed.

•	Targeted Marketing: Ensuring that promotional efforts (e.g., offering discounts) are directed toward actual potential buyers to avoid wasted resources.

Rationale

1.	Accuracy: It’s a balanced metric useful when both FP and FN have similar consequences, making it suitable for cases like spam detection or quality control where overall correctness matters most.

2.	Sensitivity: Focuses on minimizing false negatives, making it critical in high-stakes scenarios like medical diagnosis, where missing a positive case could lead to severe consequences.

3.	Specificity: Prioritizes minimizing false positives, ideal for cases like fraud detection or drug approvals, where incorrectly flagging negatives as positives leads to unnecessary costs or risks.

4.	Precision: Ensures the positive predictions are reliable, which is essential in contexts like targeted marketing or spam detection, where acting on false positives wastes resources or creates problems.
